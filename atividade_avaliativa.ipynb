peça um número ao usuário e informe se o numero é par ou impar

```python
numero = int(input("digite um numero: "))

#processamento

if numero % 2 == 0:
  print(f"O número {numero} é par.")
else:
  print(f"O número {numero} é ímpar.")
```

solicite 2 numeros e verifique qual é maior (ou se são iguais)

```python
#entrada
numero1 = int(input("digite um número: "))
numero2 = int(input("digite outro número: "))
#processamento
if numero1 > numero2:
  print(f"o número {numero1} é maior que o {numero2}")
elif numero1 == numero2:
  print(f"eles são iguais")
else:
  print(f"o número {numero2} é maior que o {numero1}")
```

aprovação de alunos (peça as notas de um aluno e informe se ele foi aprovado(nota >/=7), ficou de recuperação(nota >5 e <7) ou foi reprovado(nota <5)

```python
#entrada
nota = int(input("insira sua nota: "))

#processamento
if nota >= 7:
  print("parabens, você passou")
elif nota > 5 and nota < 7:
  print("voê está de recuperação")
else:
  print("você foi reprovado, melhore")
```

classificador de idade, receba a idade e classifique criança 0-12 anos, adolescente 13-17, adulto 18-59 e idoso 60+

```python
idade = int(input("digite sua idade: "))

if idade >= 0 and idade <= 12:
  print("você é criança")
elif idade >= 13 and idade <= 17:
  print("você é um adolescente")
elif idade >= 18 and idade <= 59:
  print("você ja é um adulto")
else:
  print("você é idoso")
```

positivo, negativo ou zero

```python
numero = int(input("digite um numero: "))
if numero > 0:
  print(f"o numero {numero} é positivo")
elif numero < 0:
  print(f"o numero {numero} é negativo")
else:
  print(f"é zero")
```

calculadora simples

```python
a = int(input("insira um numero: "))
b = int(input("insira um numero: "))

print("Opção 1 = soma")
print("Opção 2 = subtração")
print("Opção 3 = multiplicação")
print("Opção 4 = divisão")

opcao = int (input("digite a opção desejada: "))

if opcao == 1:
  print(f"{a + b}")
elif opcao == 2:
  print(f"{a - b}")
elif opcao == 3:
  print(f"{a * b}")
elif opcao == 4:
  print(f"{a / b}")
else:
  print("escolha uma opção válida")
```
